```python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
import matplotlib.pyplot as plt

import numpy as np
from psfutil_1d import *
```

```python
fig, axs = plt.subplots(2, 1, figsize=(9.6, 3.6), sharex=True)

for band, sigma in SIGMA.items():
    psf = psf_gaussian(sigma)
    print(band, psf.sum())
    visualize_psf(axs, psf, f"Gaussian, $\\sigma = {sigma:.4f}$")

axs[0].legend()
for ax in axs: ax.grid()
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(9.6, 3.6), sharex=True)

for band, ldp in LDP.items():
    psf = psf_single_slit(ldp)
    print(band, psf.sum())
    visualize_psf(axs, psf, rf"Single-Slit, $\lambda/D = {ldp:.3f}$")

axs[0].legend()
for ax in axs: ax.grid()
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(9.6, 3.6), sharex=True)

visualize_psf(axs, psf := psf_single_slit(LDP["H158"]), "Original (H158)")
visualize_psf(axs, psfp := pixelate_psf(psf), "Pixelated (H158)")

axs[0].legend()
for ax in axs: ax.grid()
fig.tight_layout()
plt.show()
```
